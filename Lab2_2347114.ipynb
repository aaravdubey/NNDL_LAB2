##### Implement the following activation functions in Python:
- **Step Function**
- **Sigmoid Function** (Binary and Bipolar)
- **Tanh Function**
- **ReLU Function**


```python
import numpy as np
import matplotlib.pyplot as plt

# Step function
def step_function(x):
    return np.where(x >= 0, 1, 0)

# Sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Bipolar Sigmoid function
def bipolar_sigmoid(x):
    return 2 / (1 + np.exp(-x)) - 1

# Tanh function
def tanh(x):
    return np.tanh(x)

# ReLU function
def relu(x):
    return np.maximum(0, x)

# Visualization function
def plot_activation_function(x, y, title):
    plt.plot(x, y)
    plt.title(title)
    plt.grid(True)
    plt.show()

# Generate inputs for visualization
x_values = np.linspace(-10, 10, 100)

# Visualize each activation function
plot_activation_function(x_values, step_function(x_values), 'Step Function')
plot_activation_function(x_values, sigmoid(x_values), 'Sigmoid Function')
plot_activation_function(x_values, bipolar_sigmoid(x_values), 'Bipolar Sigmoid Function')
plot_activation_function(x_values, tanh(x_values), 'Tanh Function')
plot_activation_function(x_values, relu(x_values), 'ReLU Function')
```

## 2. Implement a Simple Neural Network

- Create a simple neural network with one hidden layer using each activation function (`sigmoid`, `tanh`, and `ReLU`).
- Train the network on a binary classification task (e.g., XOR problem) using a small dataset.
- Compare the performance of the neural network with different activation functions.


```python
import numpy as np
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Dense
from sklearn.metrics import accuracy_score

# XOR dataset
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([0, 1, 1, 0])

# Function to build and train the model
def build_and_train_model(activation):
    model = Sequential()
    model.add(Dense(2, input_dim=2, activation=activation))  # Hidden layer
    model.add(Dense(1, activation='sigmoid'))  # Output layer for binary classification
    
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    model.fit(X, y, epochs=1000, verbose=0)  # Train the model
    
    y_pred = model.predict(X)
    y_pred = np.round(y_pred)  # Round predictions to 0 or 1
    print(f"Accuracy with {activation}: {accuracy_score(y, y_pred)}")

# Train and compare models with different activation functions
activations = ['sigmoid', 'tanh', 'relu']
for activation in activations:
    build_and_train_model(activation)
```

# INTERPRETATION

The code visualizes five different activation functions commonly used in neural networks. Each function's output is plotted against a range of input values, providing insights into their behavior. Below is an interpretation of each function based on their plots:

## 1. Step Function

- **Description:** The step function outputs a value of 1 for all input values greater than or equal to 0, and 0 for all values less than 0.
- **Plot Characteristics:**
  - The graph appears as a vertical jump from 0 to 1 at the origin (0, 0).
  - This function introduces non-linearity but is not differentiable at 0, which can hinder learning in gradient-based optimization methods.

## 2. Sigmoid Function

- **Description:** The sigmoid function smoothly transitions between 0 and 1. It is defined as \( \sigma(x) = \frac{1}{1 + e^{-x}} \).
- **Plot Characteristics:**
  - The curve approaches 0 as \( x \) approaches negative infinity and 1 as \( x \) approaches positive infinity.
  - It is S-shaped (sigmoidal) and has a range between 0 and 1.
  - The function saturates at both ends, leading to potential vanishing gradient issues during training.

## 3. Bipolar Sigmoid Function

- **Description:** The bipolar sigmoid function is similar to the standard sigmoid but maps input values to a range between -1 and 1. It is defined as \( b(x) = \frac{2}{1 + e^{-x}} - 1 \).
- **Plot Characteristics:**
  - The curve approaches -1 as \( x \) approaches negative infinity and 1 as \( x \) approaches positive infinity.
  - It exhibits similar saturation behavior to the standard sigmoid, but with a broader output range.

## 4. Tanh Function

- **Description:** The tanh function, or hyperbolic tangent, is defined as \( \tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}} \) and outputs values between -1 and 1.
- **Plot Characteristics:**
  - The graph is S-shaped and symmetric around the origin.
  - The function approaches -1 for large negative inputs and 1 for large positive inputs.
  - Unlike the sigmoid, the tanh function's outputs are centered around zero, which can improve convergence in some scenarios.

## 5. ReLU Function

- **Description:** The Rectified Linear Unit (ReLU) function outputs the input value directly if it is greater than 0; otherwise, it outputs 0. It is defined as \( \text{ReLU}(x) = \max(0, x) \).
- **Plot Characteristics:**
  - The graph shows a flat line at 0 for all negative inputs and a straight line with a slope of 1 for positive inputs.
  - ReLU is computationally efficient and helps mitigate the vanishing gradient problem, but it can suffer from the "dying ReLU" problem, where neurons can become inactive and only output 0.

## Conclusion

The visualizations provide a clear understanding of the behavior of each activation function. Choosing the appropriate activation function is crucial for the performance of a neural network, as it affects how well the model can learn complex patterns in the data. ReLU and tanh are often preferred for hidden layers, while sigmoid is commonly used for output layers in binary classification tasks.
